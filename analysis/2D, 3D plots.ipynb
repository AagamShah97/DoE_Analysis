```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load dataset into Pandas DataFrame
df = pd.read_csv("../data/DoE_Data_1_3.csv", names=['Experiment ID','Growth Time','Total Gas Flow Rate','Dew Point','Graphene Presence', 'Graphene Grade'])
#df
```

```python
# Flow Rate and Growth Time
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Growth Time (s)', fontsize = 15)
ax.set_ylabel('Total Gas Flow Rate (sccm)', fontsize = 15)
ax.set_title('2D Projection', fontsize = 20)
targets = ['No', 'Yes', 'Maybe']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = df['Graphene Presence'] == target
    ax.scatter(df.loc[indicesToKeep, 'Growth Time']
               , df.loc[indicesToKeep, 'Total Gas Flow Rate']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
# Dew Point and Flow Rate
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Dew Point (degC)', fontsize = 15)
ax.set_ylabel('Total Gas Flow Rate (sccm)', fontsize = 15)
ax.set_title('2D Projection', fontsize = 20)
targets = ['No', 'Yes', 'Maybe']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = df['Graphene Presence'] == target
    ax.scatter(df.loc[indicesToKeep, 'Dew Point']
               , df.loc[indicesToKeep, 'Total Gas Flow Rate']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
